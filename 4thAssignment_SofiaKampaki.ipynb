```python
platseq = 'CCGGGAGCCTCCACCGAAC'
#p2=codons(platseq)
def platipodia(platseq):
    YKeys = [key  for (key, value) in table.items() if value == 'Y']
    for k in YKeys:
        bases = list(YKeys[0])
        firsttry = codons(bases[0]+bases[1]+platseq)
        if 'Y' in firsttry:
            return firsttry
        sectry = codons(bases[0]+platseq + bases[2])
        if 'Y' in sectry:
            return sectry
        thirdtry = codons(platseq+bases[1]+bases[2])
        if 'Y' in thirdtry:
            return thirdtry
final = platipodia(platseq)
print(final)
```

```python
import re
a = {
    'M':('A','C'),
    'A':'A',
    'C':'C',
    'G':'G',
    'T':'T',
    'U':'U',
    'R':('A','G') 
}

def regex(k,b):
    string = ''
    a2=list(k)
    for x in a2:
        string = string + '[' + ''.join(a[x]) + ']' + '+'
    res= re.match(string,b)
    if res == None:
        return False
    elif b in res.group():
        return True
regex('AR','A')
```
