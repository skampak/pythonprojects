```python
kl=list(sw.keys())
def anag(w):
    for x in kl:
        if Counter(w) == Counter(x):
            yield w
            yield x
list(anag('ΚΑΝΤΕΡΑΚΗΣ'))
```

# 9η άσκηση

```python
arithmosofia={'Α': 1,
 'Β': 2,
 'Γ': 3,
 'Δ': 4,
 'Ε': 5,
 'Ζ': 7,
 'Η': 8,
 'Θ': 9,
 'Ι': 10,
 'Κ': 20,
 'Λ': 30,
 'Μ': 40,
 'Ν': 50,
 'Ξ': 60,
 'Ο': 70,
 'Π': 80,
 'Ρ': 100,
 'Σ': 200,
 'Τ': 300,
 'Υ': 400,
 'Φ': 500,
 'Χ': 600,
 'Ψ': 700,
 'Ω': 800,
 ' ' : 0}

def skor(w):
    s=0
    wl=list(w)
    for i in wl:
        s=s+arithmosofia[i]
    return s
skor('ΚΡΗΤΗ')
```
